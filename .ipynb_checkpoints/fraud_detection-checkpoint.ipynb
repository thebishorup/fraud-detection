```python
import pandas as pd

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

import seaborn as sb
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
print(df.describe())
```

```python
# Data Analysis --> check the 1 and 0 counts in Class feature 
sb.countplot(x = 'Class', data = df)
```

```python
# Choose dependednt variables "Amount & V1 - V28"
independent_variable = ['Amount'] + ['V%d' % number for number in range(1, 29)]

# Independent/target Varaible (Predicant) is "Class"
dependent_variable = 'Class'
```

```python
X = df[independent_variable]
y = df[dependent_variable]
```

```python
# splitting data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Create and the model instance and the fit the train data
lgr_model = LogisticRegression()
lgr_model.fit(X_train, y_train)
```

```python
# Predicting the output
predictions = lgr_model.predict(X_test)
```

```python
# Getting the classification report
classification_report(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

```python
# Finding the accuracy score
accuracy_score(y_test, predictions)
```
